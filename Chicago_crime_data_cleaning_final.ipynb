```python
# data transformation:
    #Date -> hour of day, day of week, day of month, week of month, month of year, week of year.



def week_of_month(date):
  first_day_of_month = date.replace(day = 1)
  week_of_month = (date - first_day_of_month).days // 7 + 1
  return week_of_month

df['Week_Of_Month_Of_Crime'] = df['Date'].apply(week_of_month)
df['Year_Of_Crime'] = df['Date'].dt.year
df['Month_Of_Crime'] = df['Date'].dt.month_name()
df['Week_Of_Crime'] = df['Date'].dt.isocalendar().week
df['Day_Of_Week_Of_Crime'] = df['Date'].dt.day_name()
df['Day_Of_Crime'] = df['Date'].dt.day
df['Hour_Of_Crime'] = df['Date'].dt.hour

#remove "year_of_crime" after cleaning if not req as year is already present
```

```python
# Changing data type:
    #FBI code
    #IUCR
    #all objects
for i in df.columns:
  if df[i].dtype == 'object':
    df[i] = df[i].astype('string')
```

```python
##Using rapidfuzz to find the similar items to clean the data -- misspelled data

!pip install rapidfuzz
from rapidfuzz import process, fuzz
```

```python
# functions to find the similar values
def group_similar_values(data, column):
    unique_values = data[column].unique()
    groups = []
    groups_dict = {}
    seen = set()

    for value in unique_values:
        if value not in seen:
            # Find matches within the similarity threshold
            matches = process.extract(value, unique_values, scorer=fuzz.ratio, limit=None)
            similar = [match[0] for match in matches if match[1] >= SIMILARITY_THRESHOLD]
            groups.append(similar)
            groups_dict[value] = similar
            seen.update(similar)

    return groups_dict
    
SIMILARITY_THRESHOLD = 85
```
